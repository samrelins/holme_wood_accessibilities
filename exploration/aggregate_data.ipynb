## Box Plots of Key Statistics:

```python
service_names = {
    "5000Emp": "Employment Sites - Large",
    "SS": "Secondary Schools",
    "FE": "Further Education",
    "GP": "GP",
    "Hosp": "Hospitals",
    "Food": "Food Stores",
}

observation_names = ["t", "15n", "30n", "15pct", "30pct"]

for service_code, service_name in service_names.items():
    display(HTML(f"<br><h2>{service_name}:<h2>"))
    for observation in observation_names:
        plot_image = compare_destination_features(
            lsoa_groups=[holme_wood_lsoas, brad_imd1_lsoas],
            group_names=["Holme Wood", "Brad IMD 1"],
            service=service_code,
            observation=observation
        )
        display(Image(plot_image))
```
